```python
import numpy as np
import interaction_constants

from qubit_system.geometry.regular_lattice_1d import RegularLattice1D
from qubit_system.geometry.regular_lattice_2d import RegularLattice2D
from qubit_system.geometry.regular_lattice_3d import RegularLattice3D
from qubit_system.qubit_system_classes import StaticQubitSystem
```

```python
import matplotlib.pyplot as plt
plt.rc('text', usetex=False)
plt.rc('font', family="serif", serif="CMU")
```

```python
N_RYD = 50
C6 = interaction_constants.get_C6(N_RYD)

LATTICE_SPACING = 4e-6

print(f"C6: {C6:.3e}")
print(f"Characteristic V: {C6 / (LATTICE_SPACING ** 6):.3e} Hz")
```

# Effect of Dimentionality on Energy Level Crossings

## 8 qubits


#### 1D

```python
s_qs_1d = StaticQubitSystem(
    N=8, V=C6,
    geometry=RegularLattice1D(LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-2e6, 10e6, 3)
)
s_qs_1d.plot_detuning_energy_levels(False, show=True, savefig_name="d_s_qs_8_+V_1d.png")

s_qs_1d = StaticQubitSystem(
    N=8, V=-C6,
    geometry=RegularLattice1D(LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-10e6, 2e6, 3)
)
s_qs_1d.plot_detuning_energy_levels(False, savefig_name="d_s_qs_8_-V_1d.png")
```

#### 2D. $4 \times 2$

```python
s_qs_2d = StaticQubitSystem(
    N=8, V=C6,
    geometry=RegularLattice2D((4, 2), spacing=LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-2e6, 10e6, 3)
)
s_qs_2d.plot_detuning_energy_levels(False, show=True, savefig_name="d_s_qs_8_+V_2d.png")

s_qs_2d = StaticQubitSystem(
    N=8, V=-C6,
    geometry=RegularLattice2D((4, 2), spacing=LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-10e6, 2e6, 3)
)
s_qs_2d.plot_detuning_energy_levels(False, savefig_name="d_s_qs_8_-V_2d.png")
```

#### 3D. $2 \times 2 \times 2$


```python
s_qs_3d = StaticQubitSystem(
    N=8, V=C6,
    geometry=RegularLattice3D((2, 2, 2), spacing=LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-2e6, 10e6, 3)
)
s_qs_3d.plot_detuning_energy_levels(False, show=True, savefig_name="d_s_qs_8_+V_3d.png")

s_qs_3d = StaticQubitSystem(
    N=8, V=-C6,
    geometry=RegularLattice3D((2, 2, 2), spacing=LATTICE_SPACING),
    Omega=0, Delta=np.linspace(-10e6, 2e6, 3)
)
s_qs_3d.plot_detuning_energy_levels(False, savefig_name="d_s_qs_8_-V_3d.png")


```
